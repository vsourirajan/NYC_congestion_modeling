```python
import pandas as pd
import numpy as np
```

```python
#load dataset
fname = "Hourly_Traffic_on_Metropolitan_Transportation_Authority__MTA__Bridges_and_Tunnels__Beginning_2010_20240319.csv"
df = pd.read_csv(fname)
df.head()
```

```python
#filter for data of cars entered NYC through the Queens Midtown Tunnel (Plaza ID = 7)
df = df[df['Plaza ID'] == 7]
df = df[df['Direction'] == 'I'].drop(columns=['Plaza ID', 'Direction'])
df["Total"] = df['# Vehicles - VToll'] + df['# Vehicles - E-ZPass']
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day_name()
df.tail()
```

```python
#group data and hour and convert to string
df['Hour'] = pd.to_timedelta(df['Hour'], unit='h')
df['Date'] = df['Date'] + df['Hour']
df = df.drop(columns=['Hour', "# Vehicles - E-ZPass", "# Vehicles - VToll"])
df['Date'] = df['Date'].dt.strftime('%Y%m%d%H')
df.head()
```

```python
#flip all the rows and then flip every group of 24 rows
df = df.loc[::-1]
df = df.groupby(np.arange(len(df))//24).apply(lambda x: x.iloc[::-1]).reset_index(drop=True)
df.head()
```

```python
#export df to a csv file
df.to_csv('Queens_Midtown_Tunnel_OLD.csv', index=False)
```
